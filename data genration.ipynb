# airport 

```python
import pandas as pd
from timezonefinder import TimezoneFinder

# Load the dataset containing airports
df_airports = pd.read_csv('https://ourairports.com/data/airports.csv')

# Filter out rows with NaN 'iata_code'
df_airports = df_airports.dropna(subset=['iata_code'])

# Filter the dataset to include only small and medium airports
df_airports = df_airports[df_airports['type'].isin(['small_airport', 'medium_airport', 'large_airport'])]

# Select only relevant columns
df_airports = df_airports[['name', 'iata_code', 'municipality', 'latitude_deg', 'longitude_deg', 'iso_country']]

# Rename columns to match the database schema
df_airports.columns = ['Name', 'IATA', 'Location', 'Latitude', 'Longitude', 'Country']

# Initialize TimezoneFinder
tf = TimezoneFinder()

# Get timezone for each airport based on latitude and longitude
df_airports['TimeZone'] = df_airports.apply(lambda row: tf.timezone_at(lng=row['Longitude'], lat=row['Latitude']), axis=1)


# Save the data as a CSV file
airport_data=df_airports
```

# employee

```python
import pandas as pd
from faker import Faker
import random

# Initialize Faker
fake = Faker()

# Function to generate random phone number
def generate_phone_number():
    return fake.phone_number()

# Function to generate random address
def generate_address():
    return fake.address()

# Function to generate salary based on position
def generate_salary(position):
    if position == 'Pilot':
        return round(random.randint(70000, 150000), -3)  # Round to nearest thousand
    elif position == 'Hostess':
        return round(random.randint(30000, 60000), -3)  # Round to nearest thousand
    else:
        return round(random.randint(25000, 40000), -3)  # Round to nearest thousand
# Generate DataFrame
positions = ['Pilot', 'Hostess', 'Assistant']
data = {
    'Name': [fake.name() for _ in range(10000)],
    'Position': [random.choice(positions) for _ in range(10000)],
    'Phone': [generate_phone_number() for _ in range(10000)],
    'Address': [generate_address() for _ in range(10000)],
    'gender': [random.choice(['M', 'F']) for _ in range(10000)]
}

# Add salary based on position
data['Salary'] = [generate_salary(pos) for pos in data['Position']]

# Create DataFrame
employee_data = pd.DataFrame(data)
employee_data.index=range(1,10001)

# Print DataFrame
```

```python
import pandas as pd
import numpy as np
from faker import Faker
import random

faker = Faker()

# Define the number of flights to generate
num_flights = 1000

# Generate FlightID
flight_ids = range(1, num_flights + 1)

# Randomly select airline IDs
airline_ids = random.choices(airline_data['ICAO'].values, k=num_flights)

# Randomly select departure and arrival airport IDs
departure_airport_ids = random.choices(airport_data['IATA'].values, k=num_flights)
arrival_airport_ids = random.choices(airport_data['IATA'].values, k=num_flights)

# Generate random departure and arrival datetimes
departure_datetimes = [faker.date_time_between(start_date='-30d', end_date='+30d') for _ in range(num_flights)]
arrival_datetimes = [departure + pd.Timedelta(hours=random.randint(1, 24)) for departure in departure_datetimes]

# Randomly select aircraft IDs
aircraft_ids = random.choices(aircraft_data['aircraftId'].values, k=num_flights)

# Generate random flight durations (in hours)
flight_durations = [random.randint(1, 12) for _ in range(num_flights)]

# Randomly generate available seats
available_seats = [random.randint(50, 300) for _ in range(num_flights)]

# Randomly select crew IDs
crew_ids = random.choices(crew_data['CrewID'].values, k=num_flights)

# Create DataFrame for FlightDetails
flight_details_data = {
    'FlightID': flight_ids,
    'airline_ID': airline_ids,
    'DepartureAirportID': departure_airport_ids,
    'ArrivalAirportID': arrival_airport_ids,
    'DepartureDateTime': departure_datetimes,
    'ArrivalDateTime': arrival_datetimes,
    'aircraftId': aircraft_ids,
    'FlightDuration': flight_durations,
    'AvailableSeats': available_seats,
    'CrewID': crew_ids
}

df_flight_details = pd.DataFrame(flight_details_data)

# Save the data to a CSV file
df_flight_details.to_csv('flight_details.csv', index=False)
```

# customer 

```python
import pandas as pd
from faker import Faker
import random

# Initialize Faker
fake = Faker()

# Function to generate random date of birth
def generate_date_of_birth():
    return fake.date_of_birth(minimum_age=18, maximum_age=90).strftime('%Y-%m-%d')

# Function to generate random passport number
def generate_passport_number():
    return fake.random_number(digits=9)

# Function to generate random email
def generate_email(name):
    return f"{name.lower().replace(' ', '_')}@example.com"

# Function to generate random phone number
def generate_phone_number():
    return fake.phone_number()

# Function to generate random customer type
def generate_customer_type():
    return random.choice(['person', 'company'])

# Function to generate random gender
def generate_gender():
    return random.choice(['Male', 'Female'])

# Generate DataFrame
data = {
    'customerID': range(1, 20001),
    'Name': [fake.name() for _ in range(20000)],
    'Gender': [generate_gender() for _ in range(20000)],
    'DateOfBirth': [generate_date_of_birth() if generate_customer_type() == 'person' else None for _ in range(20000)],
    'Nationality': [fake.country_code(representation="alpha-2") for _ in range(20000)],
    'PassportNumber': [generate_passport_number() if generate_customer_type() == 'person' else None for _ in range(20000)],
    'Email': [generate_email(name) for name in [fake.name() for _ in range(20000)]],
    'PhoneNumber': [generate_phone_number() for _ in range(20000)],
    'type': [generate_customer_type() for _ in range(20000)]
}

# Create DataFrame
df_customer = pd.DataFrame(data)

# If type is company, set specific fields to None
for i in range(len(df_customer)):
    if df_customer.loc[i, 'type'] == 'company':
        df_customer.loc[i, 'PassportNumber'] = None
        df_customer.loc[i, 'DateOfBirth'] = None
        df_customer.loc[i, 'Gender'] = None

# Print DataFrame
df_customer
```

```python
import pandas as pd
import numpy as np

# Define the number of bookings to generate
num_bookings = 200000

# Generate random data for booking table
booking_data = {
    'ReservationID': range(1, num_bookings + 1),
    'TransactionID': np.random.choice(transaction_data['TransactionID'], size=num_bookings),
    'CustomerID': np.random.choice(df_customer['customerID'], size=num_bookings),
    'FlightID': np.random.choice(df_flight_details['FlightID'], size=num_bookings),
    'Type': np.random.choice(['confirm', 'cancel'], size=num_bookings)
}

# Create DataFrame for booking
df_booking = pd.DataFrame(booking_data)

# Generate random data for booking info table
reservation_ids = np.random.choice(df_booking['ReservationID'], size=num_bookings * 9)
reservation_counts = np.minimum(np.bincount(reservation_ids), 9)

# Repeat each reservation ID the appropriate number of times
repeated_reservation_ids = []
for r_id, count in zip(reservation_ids, reservation_counts):
    repeated_reservation_ids.extend([r_id] * count)

booking_info_data = {
    'ReservationID': repeated_reservation_ids,
    'PassengerID': np.random.choice(df_customer['customerID'], size=len(repeated_reservation_ids)),
    'SeatID': np.random.choice(seat_data['SeatID'], size=len(repeated_reservation_ids))
}

# Create DataFrame for booking info
df_booking_info = pd.DataFrame(booking_info_data)

# Save the data to CSV files
df_booking.to_csv('booking.csv', index=False)
df_booking_info.to_csv('booking_info.csv', index=False)
```
